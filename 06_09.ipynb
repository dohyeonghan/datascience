```python
import pandas as pd

bank_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/데이터전처리/bank.csv', sep =',')
bank_df.head()
```

```python
bank_df.tail(10)
```

```python
# 행, 열
print(bank_df.shape)
```

```python
# 데이터형
print(bank_df.dtypes)
```

```python
# 결손값 확인
print(bank_df.isnull().any(axis=1))
print(bank_df.isnull().any(axis=0))
```

```python
# 결측값들의 개수 합
print(bank_df.isnull().sum(axis=1))
print(bank_df.isnull().sum(axis=0))
```

```python
# 결측값들의 합을 정렬(많은 순서대로)
print(bank_df.isnull().sum(axis=1).sort_values(ascending=False))
print(bank_df.isnull().sum(axis=0).sort_values(ascending=False))
```

```python
# 데이터형이 수치인 항목의 통계량 계산
bank_df.describe()
```

```python
# 문자열형만 통계값
bank_df.describe(include=['object'])
```
